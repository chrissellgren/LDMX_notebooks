```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt


x = np.empty(1000, dtype=object)
y = [None] * 1000

for i in range(999):
    theta = (np.pi/180)*rd.randint(1,180)
    x[i] = np.cos(theta)
    y[i] = np.sin(theta)


plt.plot(x,y,linestyle='',marker='o')
```

```python
ax = plt.figure().add_subplot(projection='3d')

# setting up a parametric curve
x = [0] * 1000
y = [0] * 1000
z = [0] * 1000
for i in range(999):
    t = (np.pi/180)*rd.randint(0,360)
    theta = (np.pi/180)*rd.randint(0,90)
    x[i] = np.cos(t)*np.sin(theta)
    y[i] = np.sin(t)*np.sin(theta)
    z[i] = np.cos(theta)
    

ax.plot(x, y, z,linestyle='',marker='.')

ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_zlabel("Z label")

plt.show()
```
